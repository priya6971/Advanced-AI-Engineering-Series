Import all the required libraries

```python
from google.genai import Client
from google.colab import userdata

# Retrieve your key from the Colab Secrets vault
API_KEY = userdata.get('GEMINI_API_KEY_2')
# Initialize the Gemini Client
client = Client(api_key=API_KEY)
```

Factual Prompt

```python
# 1. The Setup: Connect to the Gemini 3 Flash engine
# We use 'gemini-3-flash-preview' which is the 2026 state-of-the-art for speed and reasoning
model_id = "models/gemini-3-flash-preview"

# 2. The Prompt: Your input 'tokens'
user_prompt = "Explain why the sky is blue in exactly one sentence."

# 3. The Inference: Trigger the generation
response = client.models.generate_content(
   model=model_id,
   contents=user_prompt
)

# 4. The Output: Display the result of the prediction loop
print(f"Gemini's Response: {response.text}")
```

Real Time Information or Private Data

```python
# The Scenario: Asking about private, real-time data
impossible_prompt = """
I placed an order yesterday. My order ID is #AX-9921.
Can you tell me the current shipping status and when it will arrive?
"""

response = client.models.generate_content(
   model="models/gemini-3-flash-preview",
   contents=impossible_prompt
)

print(f"Gemini's Response:\n{'-'*20}\n{response.text}")
```

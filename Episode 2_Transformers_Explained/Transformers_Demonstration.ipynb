Sentiment Analysis

```python
from transformers import pipeline
classifier = pipeline(
    "sentiment-analysis",
    model = "distilbert-base-uncased-finetuned-sst-2-english"
)

result = classifier(
    "I absolutely loved the new AI features introduced by OpenAI company."
)
```

```python
print(result)
```

```python
result = classifier(
    "I hate the hallucinations generated by ChatGPT."
)

print(result)
```

```python
from transformers import pipeline

summarizer = pipeline("text-generation", model="t5-small")

text = "summarize: Artificial Intelligence is transforming the industries across the world. Organizations are increasingly adopting machine learning and generative AI to automate their tasks, improve customer experience and drive innovation."

summary = summarizer(text, max_length = 20, min_length = 10)

print(summary)
```
